```python
##################################MARTA#######################################APAGAR
# De forma a tornar o c√≥digo mais eficiente, definimos j√° as brands como as brands que encontr√°mos no dataset de treino e que tamb√©m foram identificadas na 1st delivery

#I. Prepara√ß√£o do Cat√°logo (Dicion√°rio)
    #1. Filtrar: cat√°logo √© filtrado para incluir apenas as ALLOWED_BRANDS (VW, Toyota, Audi, Ford, BMW, Skoda, Opel, Vauxhall, Mercedes-Benz, Hyundai) -> Reduz o espa√ßo de procura e foca o cat√°logo nas marcas relevantes para a empresa
    #2. Padroniza√ß√£o Manual: Mercedes-Benz para Mercedes; Volkswagen para VW; Vauxhall para Opel                                                        -> Cria consist√™ncia (permite que o modelo trate todas as varia√ß√µes de Mercedes como uma √∫nica categoria) e utiliza uma nomenclatura mais curta
    #3. Adi√ß√£o Manual: Modelos espec√≠ficos (e.g., 'Shuttle', 'Caddy Maxi Life' da VW) s√£o adicionados ao cat√°logo     
    #4. Normaliza√ß√£o que vem de utils.norm   
    #5. Estruturas de Infer√™ncia: Cria√ß√£o de dicion√°rios de mapeamento: models_by_brand, brands_by_model, e unique_brand_by_model.	Prepara a l√≥gica para inferir uma marca em falta se o modelo for √∫nico (e.g., se apenas a VW tiver o modelo 'polo', ent√£o se 'polo' aparecer sem marca, infere-se 'VW')                                                                                                         -> Prepara o cat√°logo para fuzzy matching tornando a compara√ß√£o insens√≠vel a mai√∫sculas/min√∫sculas e formata√ß√£o                                 -> Garante que modelos comuns nos dados do projeto, mas ausentes no cat√°logo, s√£o reconhecidos como v√°lidos
# II. Aplica√ß√£o da Limpeza e Infer√™ncia aos Dados de Treino/Valida√ß√£o/Teste
    #1. Tratamento de Missing Values (Brand e model): 
    #Se Brand estiver em falta, tenta-se inferir a marca se o model for √∫nico no cat√°logo (unique_brand_by_model) ou se houver um fuzzy match para o modelo que aponte para uma marca √∫nica.
    # Regra especial para o modelo 'corsa' para 'opel'
    # 
    #2.Verifica√ß√£o e Corre√ß√£o da Brand:
    # Se a Brand existir, mas n√£o estiver na lista de marcas v√°lidas, tenta-se corrigir com fuzzy match contra a lista valid_brands.
    # 
    # Corre√ß√£o do model (Dado a Brand v√°lida):
    # Corre√ß√µes Manuais: Aplica regras manuais (e.g., ('hyundai', 'i2') $\to$ 'i20').
    # Modelo em Falta: Marca a linha se apenas o model estiver em falta.
    # Fuzzy Match Local: Tenta corrigir o model com fuzzy match apenas dentro dos modelos v√°lidos para a Brand j√° estabelecida (e.g., corrige 'pol' para 'polo' se a marca for VW).
    # Fuzzy Match Global e Re-infer√™ncia de Marca: Se a corre√ß√£o local falhar, tenta-se um fuzzy match global do modelo. Se este novo modelo correto pertencer a uma √∫nica marca diferente, a Brand original √© corrigida para corresponder √† marca do modelo.
    # 
    # Marca√ß√£o para Revis√£o: 
    # Se nenhuma regra de corre√ß√£o ou infer√™ncia for bem-sucedida, a linha √© marcada com bm_status = 'needs_review'.
```

```python
# Define the allowed brands
ALLOWED_BRANDS = [
    "Volkswagen","Toyota","Audi","Ford","BMW","Skoda","Opel","Vauxhall", "Mercedes-Benz","Hyundai"
]

# Filter brand_model_dic to keep only allowed brands
brand_model_dic_norm = brand_model_dic[brand_model_dic["brand"].isin(ALLOWED_BRANDS)].copy()

# Replace "Mercedes-Benz" variations with "Mercedes"
brand_model_dic_norm["brand"] = brand_model_dic_norm["brand"].astype(str).str.replace(
    r"(?i)^mercedes[\s-]*benz$", "Mercedes", regex=True
)

# VW / Opel normalization
brand_model_dic_norm['brand'] = brand_model_dic_norm['brand'].replace('Volkswagen', 'VW', regex=False)
brand_model_dic_norm['brand'] = brand_model_dic_norm['brand'].replace('Vauxhall', 'Opel', regex=False)

# Add models manually
extra_rows = [
    {'brand': 'VW', 'model': 'Shuttle'},
    {'brand': 'VW', 'model': 'Caddy Maxi Life'},
    {'brand': 'Skoda', 'model': 'yeti outdoor'}
]
brand_model_dic_norm = pd.concat([brand_model_dic_norm, pd.DataFrame(extra_rows)], ignore_index=True)

# Get standardized columns using norm
brand_model_dic_norm['Brand'] = brand_model_dic_norm['brand'].apply(utils.norm)
brand_model_dic_norm['model'] = brand_model_dic_norm['model'].apply(utils.norm)

#Replace  'Yeti' for the brand 'Skoda' with 'yeti outdoor'
brand_model_dic_norm['model'] = brand_model_dic_norm['model'].replace('Yeti', 'yeti outdoor', regex=False) #APAGAR: REVER REDUND√ÇNCIA
brand_model_dic_norm['model'] = brand_model_dic_norm['model'].replace('SLK-Class', 'SLK', regex=False)

# Standardize X_train_norm, X_val_norm, X_test_norm
for df in [X_train_norm, X_val_norm, X_test_norm]:
    df['Brand'] = df['Brand'].apply(utils.norm)
    df['model'] = df['model'].apply(utils.norm)

# Remove duplicates from the catalog
brand_model_dic_norm = brand_model_dic_norm.drop_duplicates(subset=['Brand', 'model'])
```

```python
# list of valid brands
valid_brands = sorted(brand_model_dic_norm['Brand'].dropna().unique())

# for each brand, list of models
models_by_brand = (
    brand_model_dic_norm
    .groupby('Brand')['model']
    .apply(list)
    .to_dict()
)

# for each model, list of brands
brands_by_model = (
    brand_model_dic_norm
    .groupby('model')['Brand']
    .apply(list)
    .to_dict()
)
```

```python
###APAGAR: n√£o √© necess√°rio
# print all brands_by_model that have more than 1 brand
for model, brands in brands_by_model.items():
    if len(brands) > 1:
        print(f"{model}: {brands}")
```

```python
# if a model appears in only one brand ‚Üí we can infer the brand
unique_brand_by_model = {
    m: brands[0]
    for m, brands in brands_by_model.items()
    if len(brands) == 1
}

# manual corrections of models by (Brand, model)
MANUAL_MODEL_FIXES = {
    ('hyundai', 'i2'): 'i20', #APAGAR: TENS A CERTEZA QUE ERA ESTE? √â QUE TAMB√âM ENCONTREI O DEBAIXO EM brand_from_model
    ('hyundai', 'ix2'): 'ix20',
    # APAGAR: PODEMOS ADICIONAR MAIS REGRAS AQUI
}
```

```python
def clean_brand_model_simple(df_to_clean):
    df = df_to_clean.copy()

    if 'Brand' not in df.columns or 'model' not in df.columns:
        raise ValueError("The DataFrame must contain 'Brand' and 'model' columns.")

    df['bm_status'] = 'ok'
    df['bm_note'] = ''

    all_models = brand_model_dic_norm['model'].tolist()

    def _fix_row(row):
        b = row['Brand']
        m = row['model']

        # transform NaN to empty string
        b = '' if pd.isna(b) else str(b)
        m = '' if pd.isna(m) else str(m)

        # 0) Both Brand and model empty
        if b == '' and m == '':
            row['bm_status'] = 'empty'
            return row

        # 1) Brand missing but model filled
        if b == '' and m != '':
            # special rule: if model = 'corsa' ‚Üí brand = 'opel'
            if m == 'corsa':
                row['Brand'] = 'opel'
                row['bm_status'] = 'brand_inferred'
                row['bm_note'] = "Brand set to 'opel' because model='corsa'"
                return row

            # first: if model is unique ‚Üí infer brand
            if m in unique_brand_by_model:
                new_b = unique_brand_by_model[m]
                row['Brand'] = new_b
                row['bm_status'] = 'brand_inferred'
                row['bm_note'] = f"brand inferred from model '{m}'"
                return row

            # otherwise: fuzzy global match on model
            matches = utils.get_best_match(m, all_models)
            if len(matches) == 1:
                matched_model = matches[0]
                brands = brands_by_model.get(matched_model, [])
                if len(brands) == 1:
                    row['Brand'] = brands[0]
                    row['model'] = matched_model
                    row['bm_status'] = 'brand_inferred'
                    row['bm_note'] = f"brand inferred from fuzzy model match '{matched_model}'"
                else:
                    row['model'] = matched_model
                    row['bm_status'] = 'needs_review'
                    row['bm_note'] = f"model corrected to '{matched_model}', ambiguous brand"
            else:
                row['bm_status'] = 'needs_review'
                row['bm_note'] = "brand missing and model without clear match"
            return row

        # -------------------------
        # 2) Brand filled: ensure it is valid or correct it
        # -------------------------
        if b not in valid_brands:
            b_matches = utils.get_best_match(b, valid_brands)
            if len(b_matches) == 1:
                new_b = b_matches[0]
                row['Brand'] = new_b
                row['bm_status'] = 'brand_corrected'
                row['bm_note'] = f"brand corrected from '{b}' to '{new_b}'"
                b = new_b
            else:
                row['bm_status'] = 'needs_review'
                row['bm_note'] = f"invalid brand: '{b}'"
                return row

        # From here on, b is a valid Brand

        # Manual corrections of model by (Brand, model)
        key = (b, m)
        if key in MANUAL_MODEL_FIXES:
            new_m = MANUAL_MODEL_FIXES[key]
            row['model'] = new_m
            row['bm_status'] = 'model_corrected'
            row['bm_note'] = f"model manually corrected from '{m}' to '{new_m}'"
            m = new_m  # update for the rest of the logic

        # 3) model missing
        if m == '':
            row['bm_status'] = 'no_model'
            row['bm_note'] = "model missing"
            return row

        # -------------------------
        # 3a) exact model within the brand
        # -------------------------
        brand_models = models_by_brand.get(b, [])
        if m in brand_models:
            # perfect match
            return row

        # -------------------------
        # 3b) fuzzy match of model *within the brand*
        #  ->for example: 'pol' should go to 'polo' if it is VW, etc.
        # -------------------------
        if brand_models:
            m_matches = utils.get_best_match(m, brand_models)
            if len(m_matches) == 1:
                new_m = m_matches[0]
                row['model'] = new_m
                row['bm_status'] = 'model_corrected'
                row['bm_note'] = f"model corrected from '{m}' to '{new_m}' within the brand '{b}'"
                return row

        # -------------------------
        # 3c) global fuzzy match of model
        #  -> if this implies another brand, we also correct Brand
        # -------------------------
        m_matches_global = utils.get_best_match(m, all_models)
        if len(m_matches_global) == 1:
            new_m = m_matches_global[0]
            # check possible brands for this model
            brands_for_m = brands_by_model.get(new_m, [])
            if len(brands_for_m) == 1 and brands_for_m[0] != b:
                # trust the model and adjust brand
                new_b = brands_for_m[0]
                row['Brand'] = new_b
                row['model'] = new_m
                row['bm_status'] = 'brand_from_model'
                row['bm_note'] = f"brand changed from '{b}' to '{new_b}' because of model '{new_m}'"
            else:
                row['model'] = new_m
                row['bm_status'] = 'model_corrected'
                row['bm_note'] = f"model corrected globally from '{m}' to '{new_m}'"
            return row

        # -------------------------
        # 3d) if we still haven't caught anything ‚Üí strange
        # -------------------------
        row['bm_status'] = 'needs_review'
        row['bm_note'] = f"combination Brand='{b}', model='{m}' has no acceptable fuzzy match"
        return row

    df = df.apply(_fix_row, axis=1)
    return df
```

```python
#####################################FILIPA#######################################APAGAR
# ENGINESIZE √â A FEATURE NUM√âRICA COM MAIOR CORRELA√á√ÉO COM O PRICE (0.62)

#A dist√¢ncia √© medida numa √∫nica dimens√£o (engineSize), e a dist√¢ncia absoluta em si j√° funciona como uma m√©trica de similaridade. O scaling s√≥ √© necess√°rio quando o KNNImputer usa m√∫ltiplas features com diferentes escalas
#Como s√≥ estamos a usar 1 √∫nica vari√°vel num√©rica (engineSize) para calcular a similaridade, √© muito mais eficiente e direto calcular as dist√¢ncias absolutas e selecionar os K √≠ndices mais pequenos
```

```python
K_NEIGHBORS = 5 

# This function imputes the 'model' for rows with 'needs_review' status based on the mode of the models of the KNN in engineSize within the same Brand and, if possible, the same 'model' prefix 
def impute_model_by_engine_size(df_in):
    df = df_in.copy()
    
    #Identify rows that need imputation and clean reference rows
    mask_needs_review = df['bm_status'] == 'needs_review'
    df_needs_review = df[mask_needs_review]
    df_reference = df[df['bm_status'] != 'needs_review'].copy()
    
    corrected_models = {}

    for car_id, row in df_needs_review.iterrows():
        current_brand = row['Brand']
        current_model = row['model'] #ex: 'x', 'a', 'i2'
        current_engine_size = row['engineSize']
        
        # skip if Brand or engineSize are missing
        if pd.isna(current_brand) or pd.isna(current_engine_size):
            continue

        # Filter references by the same Brand
        ref_by_brand = df_reference[df_reference['Brand'] == current_brand].copy()

        # For short problematic models (2 or fewer characters), try to match only those starting with the same prefix
        if len(current_model) <= 2:
            ref_by_brand = ref_by_brand[
                ref_by_brand['model'].str.startswith(current_model, na=False)
            ]
        
        ref_by_brand = ref_by_brand.dropna(subset=['engineSize'])
        if ref_by_brand.empty:
            # Could not find reference neighbors with the same Brand/same prefix/engineSize
            continue

       
        distance = np.abs(ref_by_brand['engineSize'] - current_engine_size)
        # Find the K neighbors with the closest engineSize
        nearest_neighbors = distance.nsmallest(K_NEIGHBORS).index
        
        # Calculate the Mode of the models of these neighbors
        inferred_model = ref_by_brand.loc[nearest_neighbors, 'model'].mode().iloc[0] 
        corrected_models[car_id] = inferred_model

    # Update the DataFrame with the inferred models
    for car_id, new_model in corrected_models.items():
        df.loc[car_id, 'model'] = new_model
        df.loc[car_id, 'bm_status'] = 'model_imputed_by_esize'
        df.loc[car_id, 'bm_note'] = f"Model imputed to '{new_model}' using engineSize/prefix similarity."
        
    return df
```

```python
def correct_categorical_value(input_value, valid_values, min_score=0.6, fallback_value="unknown"):
    # Start by normalizing the input to have a clean base for comparison
    normalized_input = utils.norm(input_value)

    # Use get_best_match (similarity-based function) to find the best matches
    best_matches = utils.get_best_match(normalized_input, valid_values, min_score=min_score)

    # If get_best_match returns a list of matches
    if isinstance(best_matches, list):
        if len(best_matches) == 1:
            return best_matches[0]  # unique match found
        elif len(best_matches) > 1:
            print(f"Tie detected for value '{input_value}'. Multiple equally good matches found: {best_matches}")
            return None  # no correction if tie
        else:
            return fallback_value  # no good match
    # If get_best_match returns a tuple (best_match, score)
    elif isinstance(best_matches, tuple):
        return best_matches[0]
    else:
        return fallback_value
```

```python
def missing_data(train_df, val_df=None, test_df=None):
    frames = [("Train", train_df), ("Validation", val_df), ("Test", test_df)]
    out = pd.DataFrame()
    for name, df in frames:
        if df is None:
            continue
        out[f"{name} n"] = df.isna().sum()
        out[f"{name} %"] = (df.isna().mean() * 100).round(2)
    return out
```

```python
####APAGAR
# A classifica√ß√£o de uma vari√°vel como **MAR** ou **MCAR** determina a melhor estrat√©gia de imputa√ß√£o a ser usada.
# Para as vari√°veis MAR, a probabilidade de o valor estar em falta **depende** de outras *features* observadas. A imputa√ß√£o com knnimputer, e **manter a coluna bin√°ria de *missingness*** (`is_missing`) no *pipeline* do modelo, pois o facto de o dado estar em falta √©, por si s√≥, uma informa√ß√£o preditiva.

#N√£o √© poss√≠vel provar definitivamente o MAR/MCAR, mas podemos recolher evid√™ncias para apoiar ou rejeitar a hip√≥tese de MCAR.

# Justifica√ß√£o da pequena altera√ß√£o:
# O erro principal residia na utiliza√ß√£o de inplace=True para eliminar a coluna auxiliar de missingness na vers√£o original da fun√ß√£o. 
# Isso causava uma modifica√ß√£o permanente no DataFrame de treino (X_train) a cada itera√ß√£o do loop. 
# Como resultado, as itera√ß√µes subsequentes para testar outras vari√°veis (e.g., testar se a falta de year se relaciona com as features) 
# poderiam inadvertidamente incluir colunas auxiliares de itera√ß√µes anteriores que deveriam ter sido eliminadas, ou, mais importante, 
# o DataFrame original era alterado, tornando os resultados inconsistentes e dif√≠ceis de rastrear ou replicar.

#O que faz (e j√° fazia apenas juntei price e usei copy para evitar o problema do inplace):
# A fun√ß√£o test_missingness_Xtrain executa uma an√°lise sistem√°tica para avaliar a natureza dos dados em falta (MCAR vs. MAR) 
# para uma dada vari√°vel (target) em rela√ß√£o a todas as outras features do DataFrame.

#H_0 ->MCAR: A probabilidade de um valor estar em falta na vari√°vel target √© independente dos valores de qualquer outra vari√°vel (col) no DataFrame.
#H_1 -> MAR: A probabilidade de um valor estar em falta na vari√°vel target depende dos valores de pelo menos uma outra vari√°vel (col)

#********************************************************************
#1. Cria√ß√£o do Indicador de Missings: cria uma nova coluna aux (miss_col) que √© a vers√£o bin√°ria (0 ou 1) da vari√°vel target
# O tipo de teste estat√≠stico (Teste t ou Qui-Quadrado) √© escolhido com base no tipo da vari√°vel preditora que est√° a ser comparada com o status de missing (NaN ou n√£o NaN) da vari√°vel alvo.
#2. Testes para var. num√©ricas (num_cols): Teste t-Student para Duas Amostras Independentes (ttest_ind).
#Compara a m√©dia de cada vari√°vel num√©rica (sendo que para cada s√£o retirados os valores a missing) entre os dois grupos definidos pelo indicador de missings
#ex: Testa se o valor m√©dio (a distribui√ß√£o na verdade) de mileage (sem contar com NaNs) √© estatisticamente diferente para os carros onde engineSize est√° em falta em compara√ß√£o com os carros onde engineSize est√° presente
#3. Testes (Qui-Quadrado) para var. cat.: Testa a independ√™ncia entre a distribui√ß√£o de frequ√™ncias de cada vari√°vel categ√≥rica (col) e o indicador de missing
#ex: Testa se a propor√ß√£o de Diesel, Petrol ou Hybrid √© a mesma (independente) nos carros onde transmission est√° em falta e nos carros onde transmission est√° presente. Cria uma tabela de conting√™ncia das frequ√™ncias observadas para aplicar o teste chi^2


# Porque usar mais vari√°veis do knn do que aquelas que aparecem com o teste?
#O facto de uma vari√°vel estar estatisticamente relacionada √© apenas uma indica√ß√£o. 
# O KNNImputer ou a Regress√£o s√£o mais eficazes quanto mais informa√ß√µes tiverem. Excluir vari√°veis pode reduzir a precis√£o. geralmente, mais dados levam a melhor imputa√ß√£o

# Um risco que podemos estar a correr:
# As vari√°veis com apenas 2% de NaNs (cerca de 1200 NaNs) t√™m menos poder para detetar correla√ß√µes fracas, tendendo a ser classificadas como MCAR


# Para as vari√°veis MAR, O intuito: 
# A coluna original (`mpg`) fica preenchida com valores, e o modelo de ML recebe as duas colunas: 
# o valor imputado (`mpg`) e o indicador de que o valor foi imputado (`mpg_is_missing`). O modelo de ML pode aprender que: "Se `mpg_is_missing` √© 1 (o valor foi imputado), isso significa que o pre√ßo deve ser, por exemplo, ¬£1000 a menos, mesmo depois de imputar o valor de `mpg`.


# O principal problema de imputar todas as vari√°veis (incluindo as MCAR) com um KNNImputer √© o risco de adicionar complexidade desnecess√°ria e bias, em vez de se obter o ganho preditivo pretendido.

#N√∫mero de Testes Feitos: $N_Testes = N_vars com NaNs * N_outras features$
#Para cada vari√°vel com NaNs, o c√≥digo executa um teste contra todas as outras features no DataFrame 
```

```python
# Function to Test MAR/MCAR: For a given target column with missing data, test if the "missingness" is statistically related to all other columns in the dataset

def test_missingness_Xtrain(X_train_in, target, num_cols):
    X_train = X_train_in.copy()

    # Create missingness indicator (1 = missing, 0 = not missing)
    miss_col = target + '_missing'
    X_train[miss_col] = X_train[target].isna().astype(int)
    
    results = []

    #  Numeric predictors:  t-tests
    for col in num_cols:
        if col != target and col in X_train.columns:
            group0 = X_train.loc[X_train[miss_col] == 0, col].dropna()
            group1 = X_train.loc[X_train[miss_col] == 1, col].dropna()
            if len(group0) > 1 and len(group1) > 1:
                p = ttest_ind(group0, group1, equal_var=False)[1]
                results.append((target, col, 't-test', p))

    #  Categorical predictors:  chi-square
    cat_cols = X_train.select_dtypes(exclude='number').columns
    for col in cat_cols:
        if col != miss_col:
            table = pd.crosstab(X_train[col], X_train[miss_col])
            if table.shape[0] > 1:
                _, p, _, _ = chi2_contingency(table)
                results.append((target, col, 'chi-square', p))
   
    return pd.DataFrame(results, columns=['Target', 'Variable', 'Test', 'p_value'])
```

```python
# Calling MAR/MCAR test function on all variables with missing values

# join X_train and y_train to have price available
X_train_with_price = X_train.join(y_train)

# Identify variables that have missing values
vars_with_missing = [col for col in X_train_with_price.columns if X_train_with_price[col].isna().sum() > 0 and col != 'price']
numeric_cols = X_train_with_price.select_dtypes('number').columns
all_results = []

for var in vars_with_missing:
    results = test_missingness_Xtrain(X_train_with_price, var, numeric_cols)
    sig_vars = results[results['p_value'] < 0.05]['Variable'].tolist()
    status = "MAR" if len(sig_vars) > 0 else "MCAR"
    all_results.append({
        'Target': var,
        'Missing_Count': X_train[var].isna().sum(),
        'Missing_%': X_train[var].isna().mean() * 100,
        'Significant_Variables': sig_vars,
        'Status': status
    })

# Combine into one summary DataFrame
summary = pd.DataFrame(all_results).sort_values('Missing_%', ascending=False)
summary.reset_index(drop=True, inplace=True)

pd.set_option('display.max_colwidth', None)


# Clean up columns
summary['Missing_Count'] = summary['Missing_Count'].astype(int)
summary['Missing_%'] = summary['Missing_%'].round(2)
summary['Significant_Variables'] = summary['Significant_Variables'].apply(
    lambda lst: ', '.join(lst) if isinstance(lst, list) and len(lst) > 0 else '‚Äî'
)

# Reorder + sort + reset index
summary = (
    summary[['Target', 'Missing_Count', 'Missing_%', 'Status', 'Significant_Variables']]
    .sort_values(['Missing_%', 'Target'], ascending=[False, True])
    .reset_index(drop=True)
)

print(summary.to_string(index=False))
```

```python
def compare_outlier_detection(df, columns, iqr_factor=1.5, mad_threshold=3.5, return_indices=False):

    # Just to ensure columns is a list
    if isinstance(columns, str):
        columns = [columns]
    
    results_list = []
    all_indices = {'iqr': set(), 'mad': set(), 'both': set()}

    # Iterate through each column
    for column in columns:
        if column not in df.columns:
            print(f"Column '{column}' not found in the DataFrame. Skipping.")
            continue
        
        data_series = df[column]

        ### IQR Method Calculations ###
        Q1 = data_series.quantile(0.25)
        Q3 = data_series.quantile(0.75)
        IQR = Q3-Q1
        iqr_lower_bound = Q1- (iqr_factor * IQR)
        iqr_upper_bound = Q3+ (iqr_factor * IQR)
        iqr_outliers_filter = (data_series < iqr_lower_bound) | (data_series > iqr_upper_bound)
        iqr_outlier_indices = set(df[iqr_outliers_filter].index)
        
        ### MAD Method Calculations ###
        median = data_series.median()
        mad = (data_series - median).abs().median()
        
        # Handle case where MAD is zero
        if mad == 0:
            mad_lower_bound = median
            mad_upper_bound = median
            # If MAD is zero, consider all values as outliers that are not equal to the median
            mad_outliers_filter = (data_series != median)
        else:
            mad_lower_bound = median - (mad_threshold * mad)
            mad_upper_bound = median + (mad_threshold * mad)
            mad_outliers_filter = (data_series < mad_lower_bound) | (data_series > mad_upper_bound)
        
        mad_outlier_indices = set(df[mad_outliers_filter].index)

        # Comparison Calculations
        both_indices = iqr_outlier_indices.intersection(mad_outlier_indices)
        
        column_summary = {
            'Column': column,
            'Outlier Count (IQR)': len(iqr_outlier_indices),
            'Outlier % (IQR)': round(len(iqr_outlier_indices) / len(df) * 100, 2),
            'Outlier Count (MAD)': len(mad_outlier_indices),
            'Outlier % (MAD)': round(len(mad_outlier_indices) / len(df) * 100, 2),
            'Outlier Count (Both)': len(both_indices),
            'Outlier % (Both)': round(len(both_indices) / len(df) * 100, 2),
        }
        results_list.append(column_summary)

        # Store indices if requested
        if return_indices:
            all_indices['iqr'].update(iqr_outlier_indices)
            all_indices['mad'].update(mad_outlier_indices)
                    
    if return_indices:
        all_indices['both'] = all_indices['iqr'].intersection(all_indices['mad'])
        for key in all_indices:
            all_indices[key] = sorted(list(all_indices[key]))

    summary_df = pd.DataFrame(results_list).set_index('Column')
    
    if return_indices:
        return summary_df, all_indices
        
    return summary_df
```

```python
def fit_keep_models_per_brand(train, brand_col="Brand", model_col="model",
                              min_count=20, min_freq=0.01):
    tmp = train[[brand_col, model_col]].copy()

    # Exclude 'unknown' entries
    tmp = tmp[~tmp[brand_col].isin(["unknown"])]
    tmp = tmp[~tmp[model_col].isin(["unknown"])]

    # Count combinations and calculate frequencies within each brand
    g = (tmp.groupby([brand_col, model_col], dropna=False)
           .size().rename("n").reset_index())
    g["brand_total"] = g.groupby(brand_col)["n"].transform("sum")
    g["freq"] = g["n"] / g["brand_total"]

    # Keep only sufficiently frequent combinations
    keep = g.loc[
        (g["n"] >= min_count) & (g["freq"] >= min_freq),
        [brand_col, model_col]
    ].copy()
    keep["keep"] = True

    return keep


def collapse_rare_models(df, keep_table, brand_col="Brand", model_col="model",
                         other_label="other"):
    # Save original index
    original_index = df.index

    out = df.copy()

    # Merge with keep_table (preserving all rows from df)
    out = out.merge(
        keep_table, how="left",
        on=[brand_col, model_col]
    )

    keep_mask = out["keep"].fillna(False)

    # Only replace rare models, skip if brand or model == 'unknown'
    mask_to_replace = (
        (~keep_mask) &
        (~out[brand_col].isin(["unknown", None, np.nan])) &
        (~out[model_col].isin(["unknown", None, np.nan]))
    )

    out.loc[mask_to_replace, model_col] = other_label

    # Restore original index
    out.index = original_index

    # Drop auxiliary column
    out = out.drop(columns=["keep"], errors="ignore")

    return out
```

```python
####################################FILIPA#######################################APAGAR
# Local Outlier Factor (LOF) detetar outliers multivariados
#LOF √© um algoritmo n√£o supervisionado de dete√ß√£o de anomalias que identifica outliers ao medir o qu√£o isolado 
#um ponto de dados est√° em rela√ß√£o aos seus vizinhos, uma abordagem conhecida como dete√ß√£o baseada em densidade local.

#Hip√≥tese Central (Densidade): Os outliers t√™m uma densidade local significativamente menor do que os seus vizinhos. Isto significa que o LOF √© eficaz a encontrar outliers mesmo que eles estejam dentro da "nuvem" de dados, mas numa regi√£o menos densa (chamados outliers de cluster).
#Hip√≥tese de Dist√¢ncia: O conceito de vizinhan√ßa e dist√¢ncia  √© v√°lido para o espa√ßo de features onde o algoritmo opera.
#Temos de usar scaling mas pelo menos √© N√£o-Param√©trico

#Dado na AP de instance based learning

#Ideia Central: LOF mede a densidade local de um ponto e compara-a com a densidade dos seus K vizinhos mais pr√≥ximos.
#Score:         LOF Score (negative_outlier_factor_ na implementa√ß√£o do scikit-learn) quantifica esta anomalia.
#Interpreta√ß√£o: LOF Score approx 1 indica que o ponto tem uma densidade semelhante √† dos vizinhos (√© um inlier/normal). 
#               LOF Score > 1 (ou negative_outlier_factor_ muito menor que -1) indica que o ponto est√° numa regi√£o mais esparsa do que os seus vizinhos, classificando-o como outlier.
#Par√¢metros-Chave: 1. n_neighbors (n√∫mero de vizinhos para estimar a densidade) e 
#                  2. contamination (percentagem esperada de outliers no conjunto de dados).

#o maior down side √© que  somos n√≥s a decidir o que √© um outlier (definir contamination) e n√£o o algoritmo
```

```python
from sklearn.model_selection import KFold

# KFold Target Encoding
kf = KFold(n_splits=5, shuffle=True, random_state=37)
global_mean = y_train['price'].mean()  # ‚Üê MUDAN√áA AQUI

for col in high_cardinality_cols:
    # Initialize
    X_train_processed[f'{col}_target_enc'] = 0.0
    fold_encodings = []
    
    # Out-of-fold encoding (prevents leakage)
    for train_idx, val_idx in kf.split(X_train):
        # Calculate encoding on train fold
        temp = pd.DataFrame({
            'cat': X_train.iloc[train_idx][col].values,  # ‚Üê .values
            'target': y_train.iloc[train_idx]['price'].values  # ‚Üê .values e ['price']
        })
        encoding = temp.groupby('cat')['target'].mean().to_dict()
        fold_encodings.append(encoding)
        
        # Apply to val fold
        X_train_processed.loc[X_train.index[val_idx], f'{col}_target_enc'] = (
            X_train.iloc[val_idx][col].map(encoding).fillna(global_mean)
        )
    
    # Average encoding for val/test
    all_cats = set().union(*[enc.keys() for enc in fold_encodings])
    avg_encoding = {cat: np.mean([enc.get(cat, global_mean) for enc in fold_encodings]) 
                    for cat in all_cats}
    
    # Apply to val and test
    X_val_processed[f'{col}_target_enc'] = X_val[col].map(avg_encoding).fillna(global_mean)
    X_test_processed[f'{col}_target_enc'] = X_test[col].map(avg_encoding).fillna(global_mean)

# Drop original columns
X_train_processed.drop(columns=high_cardinality_cols, inplace=True)
X_val_processed.drop(columns=high_cardinality_cols, inplace=True)
X_test_processed.drop(columns=high_cardinality_cols, inplace=True)
```

```python
for epoch in range(N_EPOCHS):

    # Treinar mais 1 √©poca
    model.fit(X_train_final_reset, y_train_reset)

    # Previs√µes
    y_train_pred = model.predict(X_train_final_reset)
    y_val_pred   = model.predict(X_val_final_reset)

    # MAE para treino e valida√ß√£o
    train_mae.append(mean_absolute_error(y_train_reset, y_train_pred))
    val_mae.append(mean_absolute_error(y_val_reset, y_val_pred))

# Plot da curva de aprendizagem
plt.figure(figsize=(10,6))
plt.plot(train_mae, label="Train MAE")
plt.plot(val_mae,   label="Validation MAE")
plt.xlabel("Epochs")
plt.ylabel("MAE")
plt.title("Learning Curve (Regression - MAE)")
plt.legend()
plt.grid(True)
plt.show()

# Melhor √©poca (onde o MAE de valida√ß√£o √© m√≠nimo)
best_epoch = np.argmin(val_mae)
print("Melhor √©poca:",
      best_epoch,
      " | MAE Val =", val_mae[best_epoch])
```

```python
final_model = MLPRegressor(hidden_layer_sizes=hidden_layers,
                           activation='relu',
                           solver='adam',
                           max_iter=best_epoch,
                           random_state=0)

final_model.fit(X_train_final_reset, y_train_reset)

print("MAE Treino:", mean_absolute_error(y_train_reset, final_model.predict(X_train_final_reset)))
print("MAE Val:",    mean_absolute_error(y_val_reset, final_model.predict(X_val_final_reset)))
```
