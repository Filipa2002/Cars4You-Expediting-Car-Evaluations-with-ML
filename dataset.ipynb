```python
from pathlib import Path
import pandas as pd

# ðŸ‘‰ ajusta para a tua pasta
folder = Path(r"C:\Users\Utilizador\.cache\kagglehub\datasets\adityadesai13\used-car-dataset-ford-and-mercedes\versions\3")

# exceÃ§Ãµes especÃ­ficas que pediste
file_to_brand_override = {
    "focus": "vw",
    "cclass": "merc"
}

def canon(b: str) -> str:
    b = b.lower()
    b = file_to_brand_override.get(b, b)      # 1) aplica overrides por ficheiro
    return b

all_dfs = []
for f in sorted(folder.glob("*.csv")):
    base = f.stem.lower()                     # nome do ficheiro sem extensÃ£o
    brand = canon(base)
    df = pd.read_csv(f)
    df.insert(0, "Brand", brand)              # adiciona a coluna na frente
    all_dfs.append(df)

cars_df = pd.concat(all_dfs, ignore_index=True)
print(cars_df.shape)
cars_df.head()
```
