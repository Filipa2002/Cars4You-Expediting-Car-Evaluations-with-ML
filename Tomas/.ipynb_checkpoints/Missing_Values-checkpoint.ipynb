### Adressing Missing values: Systematic Approach (2)

```python
# Contrary to the first approach, here the imputation method is decided case by case (variable by variable)

# Logic (Little & Rubin, 2019)
# Step 1: Evaluate pattern of missing relation to other variables (missing mecanism) 

# MCAR (Missing Completely at Random):
# Missingness independent of observed/unobserved data: safe to impute with central tendency (median/mean/mode).

# MAR (Missing At Random):
# Missingness explained by another variable (e.g., tax missing for electric cars): impute within group (conditionally) / KNN / Or MICE

# MNAR (Missing Not At Random):
# Depends on unobserved values (e.g., cars too new for tax data): possibly removal or specific enconding

# Step 2: Consider Missing Proportion relative to Dataset and Datatype
# Step 3: Apply method
```

```python
# Testing for MCAR

# (1) Little's test
# Little’s MCAR test is a multivariate test that formally evaluates whether data are
# Missing Completely at Random (MCAR) — that is, whether the probability of missingness is independent of all observed and unobserved data.
# (Little, 1988)

# from pyampute.exploration import test_mcar
# res = test_mcar(numeric_cols)
# print(res)


# (2) Alternative: Manually test: Create dummy variable for missing or not
#   2.1 Create dummy
# car_eval['mpg_missing'] = car_eval['mpg'].isna().astype(int)

# 2.2 Conduct tests

# t-tests (numeric variables)
## p > 0.05 → no significant difference → missingness likely random
## p < 0.05 → missingness depends on that variable → not MCAR
from scipy.stats import ttest_ind

for col in car_eval.select_dtypes(include='number'):
    if col not in ['mpg', 'mpg_missing']:
        group0 = car_eval.loc[car_eval['mpg_missing'] == 0, col].dropna()
        group1 = car_eval.loc[car_eval['mpg_missing'] == 1, col].dropna()
        if len(group0) > 1 and len(group1) > 1:
            p = ttest_ind(group0, group1, equal_var=False)[1]
            print(f"{col}: p = {p:.4f}")
```

```python
# Chi-square test (for categorical variables)

from scipy.stats import chi2_contingency

for col in car_eval.select_dtypes(exclude='number'):
    table = pd.crosstab(car_eval[col], car_eval['mpg_missing'])
    if table.shape[0] > 1:
        chi2, p, _, _ = chi2_contingency(table)
        print(f"{col}: p = {p:.4f}")
```

## Defining function to test MAR

```python
from scipy.stats import ttest_ind, chi2_contingency


def test_missingness_Xtrain(X_train, target, num_cols):
    """Tests if missingness in `target` depends on other variables."""
    
    # Create missingness indicator (1 = missing, 0 = not missing)
    miss_col = target + '_missing'
    X_train[miss_col] = X_train[target].isna().astype(int)
    
    results = []

    #  Numeric predictors:  t-tests
    for col in numeric_cols:
        if col != target and col in X_train.columns:
            group0 = X_train.loc[X_train[miss_col] == 0, col].dropna()
            group1 = X_train.loc[X_train[miss_col] == 1, col].dropna()
            if len(group0) > 1 and len(group1) > 1:
                p = ttest_ind(group0, group1, equal_var=False)[1]
                results.append((target, col, 't-test', p))

    #  Categorical predictors:  chi-square
    cat_cols = X_train.select_dtypes(exclude='number').columns
    for col in cat_cols:
        if col != target:
            table = pd.crosstab(X_train[col], X_train[miss_col])
            if table.shape[0] > 1:
                chi2, p, _, _ = chi2_contingency(table)
                results.append((target, col, 'chi-square', p))
    
    # Remove helper column
    X_train.drop(columns=[miss_col], inplace=True)
    
    return pd.DataFrame(results, columns=['Target', 'Variable', 'Test', 'p_value'])
```

### MAR Variables

# Mpg

Numerical, 10.40% missing, MAR: year, Brand, model, transmission, fuelType

Method: MICE

Justification: (Joel, Doorsamy, & Paul, 2025): In mixed datasets with ~10% MAR missingness and multiple predictors (numerical like year; categorical like Brand/model/transmission/fuelType), MICE's iterative chained equations model each variable conditionally on others, yielding lower RMSE/MAE (e.g., 15-20% improvement over KNN in benchmarks on tabular data with 5-15% missingness) by preserving inter-variable relationships without normality assumptions. KNN, while effective for similarity-based imputation, underperforms here due to challenges in distance calculations across mixed types, showing higher errors in 70% of simulated MAR scenarios. Data from a 2025 comparative study on 20 datasets confirms MICE's superiority for such profiles, with average RMSE reductions of 0.12-0.25 units versus KNN.

https://journals.plos.org/plosone/article?id=10.1371/journal.pone.0254720


```python
# Applying MICE

# Missing mpg values ARE  imputed using Multiple Imputation by Chained Equations (MICE), implemented via IterativeImputer from Scikit-learn. 
## This method iteratively predicts each variable with missing data using regression models built on the other variables, providing statistically consistent imputations under the MAR "assumption".

from sklearn.preprocessing import LabelEncoder
from sklearn.experimental import enable_iterative_imputer  
from sklearn.impute import IterativeImputer
import numpy as np

# 1) Encode categoricals on a copy (required to run mice)
X_train_enc = X_train.copy()
encoders = {}
for col in ['Brand', 'model', 'fuelType', 'transmission']:
    le = LabelEncoder()
    X_train_enc[col] = le.fit_transform(X_train_enc[col].astype(str))
    encoders[col] = le  

# 2) Define features (all numeric after enconding)
mice_features_mpg = [
    'mpg', 'engineSize', 'mileage', 'tax', 'year',
    'Brand', 'model', 'fuelType', 'transmission'
]

# 3) Fit MICE on the ENCODED dataframe
mice_imputer_mpg = IterativeImputer(max_iter=10, random_state=42)
mice_imputer_mpg.fit(X_train_enc[mice_features_mpg])

# 4) Transform ENCODED data
X_train_imputed_block = mice_imputer_mpg.transform(X_train_enc[mice_features_mpg])

# 5) Write back ONLY the imputed mpg into the ORIGINAL X_train
mpg_col_idx = mice_features_mpg.index('mpg')
X_train.loc[:, 'mpg'] = X_train_imputed_block[:, mpg_col_idx]
```

## PaintQuality% 

Numerical, 2.03% missing, MAR: model

Method: Grouped simple imputation (Median)

Justification: (Austin, White, Lee, & van Buuren, 2021): For low-missingness (<5%) numerical MAR with a single categorical predictor, grouped mean imputation conditions directly on the dependency (in this case model), achieving near-zero bias in distribution preservation per group (e.g., variance inflation <5% in simulations) and outperforming MICE in computational efficiency without loss in accuracy for sparse cases. Benchmarks on categorical-driven MAR show simple grouped methods match advanced ones in 90% of low-rate scenarios, as over-modeling introduces unnecessary noise.

```python
# Step 1: Compute median paintQuality% per model group
group_median = X_train.groupby('model')['paintQuality%'].median()

# Step 2: Fill missing values with group median
# transform to broadcast the median back to each row, then fill NaNs
X_train['paintQuality%_grouped'] = X_train.groupby('model')['paintQuality%'].transform(lambda x: x.fillna(x.median()))

# Step 3: Handle groups with all NaNs (fallback to overall median)
overall_median = X_train['paintQuality%'].median()
X_train['paintQuality%_grouped'] = X_train['paintQuality%_grouped'].fillna(overall_median)
```

# Transmission

Categorical, 2.03% missing, MAR; significant: Brand

Method: Groupes Simple imputation (Mode)

Justification: Analougous to PaintQuality%



```python
# Grouped mode by brand

# Ensure string dtype 
X_train['transmission'] = X_train['transmission'].astype(str)
X_train['Brand'] = X_train['Brand'].astype(str)

# Computed within-Brand mode 
brand_mode = (
    X_train.groupby('Brand')['transmission']
           .agg(lambda s: s.mode().iloc[0] if not s.mode().empty else np.nan)
)

# Fill (in case of) missing with Brand-specific mode
mask_missing = X_train['transmission'].isna() | (X_train['transmission'].str.lower() == 'nan')
X_train.loc[mask_missing, 'transmission'] = X_train.loc[mask_missing, 'Brand'].map(brand_mode)

# Fallback to global mode for brands with no mode (or any str 'nan' leftovers)
global_mode_series = X_train['transmission'][~(X_train['transmission'].str.lower()=='nan')].mode()
global_mode_val = global_mode_series.iloc[0] if not global_mode_series.empty else 'Unknown'
X_train['transmission'] = X_train['transmission'].replace('nan', np.nan).fillna(global_mode_val)
```

# Fueltype

Categorical, 1.96% missing, MAR: Brand, model

Method: Grouped Simple imputation (Mode)

Justification:  Two categorical predictors and relatively low missingness

```python
# Grouped imputation: Median


# Ensure correct dtype (to be removed)
X_train['fuelType'] = X_train['fuelType'].astype(str)
X_train['Brand'] = X_train['Brand'].astype(str)
X_train['model'] = X_train['model'].astype(str)

# Mode by Brand + Model (most granular)
brand_model_mode = (
    X_train.groupby(['Brand', 'model'])['fuelType']
           .agg(lambda s: s.mode().iloc[0] if not s.mode().empty else np.nan)
)

# Mode by Brand only (fallback if models with no valid mode)
brand_mode = (
    X_train.groupby('Brand')['fuelType']
           .agg(lambda s: s.mode().iloc[0] if not s.mode().empty else np.nan)
)

# Global mode ( fallback)
global_mode = (
    X_train['fuelType'][~X_train['fuelType'].isin(['nan', 'None', 'NaN'])]
    .mode().iloc[0]
)

# Fill in order of specificity (cover each scenario)
mask_missing = X_train['fuelType'].isna() | (X_train['fuelType'].str.lower().isin(['nan', 'none']))
for idx in X_train[mask_missing].index:
    brand, model = X_train.loc[idx, ['Brand', 'model']]
    val = brand_model_mode.get((brand, model), np.nan)
    if pd.isna(val):
        val = brand_mode.get(brand, np.nan)
    if pd.isna(val):
        val = global_mode
    X_train.at[idx, 'fuelType'] = val
```

# Mileage

Numerical, 1.89% missing, MAR; : model, transmission

Method: Grouped Simple imputation

Justification:  Two categorical predictors and low missingness

```python
# Grouped Imputation: Median

# Ensuring correct dtype (to be removed)
X_train['mileage'] = pd.to_numeric(X_train['mileage'], errors='coerce')
X_train['model'] = X_train['model'].astype(str)
X_train['transmission'] = X_train['transmission'].astype(str)

# Compute group medians by (model, transmission)
group_median_mileage = (
    X_train.groupby(['model', 'transmission'])['mileage']
    .median()
)

# Fill missing values with grouped medians
def impute_grouped_median(row):
    if pd.isna(row['mileage']):
        return group_median_mileage.get((row['model'], row['transmission']), np.nan)
    else:
        return row['mileage']

X_train['mileage_grouped'] = X_train.apply(impute_grouped_median, axis=1)

# Fallback — use overall median if still missing
overall_median_mileage = X_train['mileage'].median()
X_train['mileage_grouped'] = X_train['mileage_grouped'].fillna(overall_median_mileage)

# Replace original column and clean up
X_train['mileage'] = X_train['mileage_grouped']
X_train.drop(columns=['mileage_grouped'], inplace=True)
```
